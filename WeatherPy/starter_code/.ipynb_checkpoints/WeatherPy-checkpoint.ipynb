```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import random
import numpy as np
from scipy.stats import linregress
from pprint import pprint
import datetime as dt
import requests
import gmaps
import os
import json
import time

# Importing API key
from api_keys import g_key
from api_keys import weather_api_key
```

```python
# Cities.csv file
# Converting it into a DataFrame
# Printing the DataFrame to view the columns and data
output_file = "output_data/cities.csv"
output_file_df = pd.read_csv("../output_data/cities.csv")
output_file_df
```

```python
# Setting a range of latitude and longitude values
latitude_range = (-90, 90)
longitude_range = (-180, 180)
```

## Generate Cities List

```python
# Creating empty list of cities and an empty listof lat and long values
cities = []
lat_and_long_values = []

# Creating random latitude and longitude combinations
latitude_values = np.random.uniform(latitude_range[0], latitude_range[1], size=1500)
longitude_values = np.random.uniform(longitude_range[0], longitude_range[1], size=1500)
lat_and_long_values = zip(latitude_values, longitude_values)

# Using a "for loop" to identify hte neaest city for each latitude value and longitude value
for lat_and_long_value in lat_and_long_values:
    city = citipy.nearest_city(lat_and_long_value[0], lat_and_long_value[1]).city_name
    
    if city not in cities:
        cities.append(city)

# Printing the number of cities
len(cities)
```

```python
# Printing the cities
print(cities)
```

```python
# Adding today's date in my charts
date = dt.date.today().strftime("%m/%d/%Y")
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Base URL
url = "https://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Building query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

```python
# Creating empty lists that will hold my variables
city_name = []
lat = []
long = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []
```

```python
# Initializing count
count = 1
```

```python
# Using "for loop" to make API calls
for city in cities:
    try:
        response = requests.get(query_url + city).json()
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        long.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        
        print(f"{count} | {city}") 
        
        count += 1
        
    except:
        print(f"No city found!")
        continue
print(f"Done!")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Creating a dictionary that will hold all the variables defined in the previous section

cities_dictionary = {
    "City": city_name,
    "Cloudiness":cloudiness, 
    "Country":country,
    "Date":date, 
    "Humidity": humidity,
    "Latitude":lat, 
    "Longitude":long, 
    "Maximum_Temperature": max_temp,
    "Wind_Speed":wind_speed
}
```

```python
# Creating DataFrame
cities_data = pd.DataFrame.from_dict(cities_dictionary, orient='index')
cities_data = cities_data.transpose()
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Creating Nothern and Southern Hemisphere DataFrames
northern_hemisphere_df = cities_data.loc[pd.to_numeric(cities_data["Latitude"]).astype(float) > 0, :]
southern_hemisphere_df = cities_data.loc[pd.to_numeric(cities_data["Latitude"]).astype(float) < 0, :]
```

```python
# Plotting Northern Hemisphere - Max Temperature vs. Latitude Linear Regression
x_values = pd.to_numeric(northern_hemisphere_df['Latitude']).astype(float)
y_values = pd.to_numeric(northern_hemisphere_df['Maximum_Temperature']).astype(float)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(f"Regression line equation is {line_eq}")
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Maximum Temperature')
plt.title(f'Northern Hemisphere - Max Temp vs. Latitude Linear Regression')
print(f"The r-squared is {rvalue}")
print("The regression displays a negative correlation.")
print("The temperature decreases as you move away from the equator.")
plt.show()
```
